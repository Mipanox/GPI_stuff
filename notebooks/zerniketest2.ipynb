## RMS evaluation
Conversion factors for rms level <--> normalized coefficients defined in the class

```python
from sympy import *
```

```python
r = Symbol('r') # rho
t = Symbol('t') # theta
```

#### Expressions for the first 15 modes

```python
zer01 = 1*(cos(t)**2 + sin(t)**2)
zer02 = 2*r*cos(t)
zer03 = 2*r*sin(t)
zer04 = np.sqrt(3)*(2*r**2 - 1)
zer05 = np.sqrt(6)*r**2 * sin(2*t)
zer06 = np.sqrt(6)*r**2*cos(2*t)
zer07 = np.sqrt(8)*(3*r**2 - 2) * r*sin(t)
zer08 = np.sqrt(8)*(3*r**2 - 2) * r*cos(t)
zer09 = np.sqrt(8)*r**3*sin(3*t)
zer10 = np.sqrt(8)*r**3*cos(3*t)
zer11 = np.sqrt(5)*(1-6*r**2 + 6*r**4)
zer12 = np.sqrt(10)*(4*r**2-3)*r**2 * cos(2*t)
zer13 = np.sqrt(10)*(4*r**2-3)*r**2 * sin(2*t)
zer14 = np.sqrt(10)*r**4 * cos(4*t)
zer15 = np.sqrt(10)*r**4 * sin(4*t)
```

#### rms

```python
## integrals
int01 = integrate(zer01**2 * r, (r,0,1), (t,0,2*pi))
int02 = integrate(zer02**2 * r, (r,0,1), (t,0,2*pi))
int03 = integrate(zer03**2 * r, (r,0,1), (t,0,2*pi))
int04 = integrate(zer04**2 * r, (r,0,1), (t,0,2*pi))
int05 = integrate(zer05**2 * r, (r,0,1), (t,0,2*pi))
int06 = integrate(zer06**2 * r, (r,0,1), (t,0,2*pi))
int07 = integrate(zer07**2 * r, (r,0,1), (t,0,2*pi))
int08 = integrate(zer08**2 * r, (r,0,1), (t,0,2*pi))
int09 = integrate(zer09**2 * r, (r,0,1), (t,0,2*pi))
int10 = integrate(zer10**2 * r, (r,0,1), (t,0,2*pi))
int11 = integrate(zer11**2 * r, (r,0,1), (t,0,2*pi))
int12 = integrate(zer12**2 * r, (r,0,1), (t,0,2*pi))
int13 = integrate(zer13**2 * r, (r,0,1), (t,0,2*pi))
int14 = integrate(zer14**2 * r, (r,0,1), (t,0,2*pi))
int15 = integrate(zer15**2 * r, (r,0,1), (t,0,2*pi))

## rms
domain = integrate(r, (r,0,1), (t,0,2*pi))

rms = []
rms.append(sqrt(int01 / domain))
rms.append(sqrt(int02 / domain))
rms.append(sqrt(int03 / domain))
rms.append(sqrt(int04 / domain))
rms.append(sqrt(int05 / domain))
rms.append(sqrt(int06 / domain))
rms.append(sqrt(int07 / domain))
rms.append(sqrt(int08 / domain))
rms.append(sqrt(int09 / domain))
rms.append(sqrt(int10 / domain))
rms.append(sqrt(int11 / domain))
rms.append(sqrt(int12 / domain))
rms.append(sqrt(int13 / domain))
rms.append(sqrt(int14 / domain))
rms.append(sqrt(int15 / domain))
```

```python
for i in range(len(rms)):
    print 'The rms of the {0}th Zernike mode is {1}'.format(i+1,rms[i])
```

Thus prove that the normalization choice corresponds exactly to that the rms levels be equal to the coefficients
